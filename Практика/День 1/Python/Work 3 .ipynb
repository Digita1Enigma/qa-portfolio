```python
def is_prime(num):
    return len([i for i in range(1, num+1) if num % i == 0]) == 2
n = int(input())
print(is_prime(n))
```

```python
def get_next_prime(num):
    num += 1
    for i in range(2, num):
        if num % i == 0:
            return get_next_prime(num)
    return num
n = int(input())
print(get_next_prime(n))
```

```python
def is_password_good(password):
    upp = [i for i in password if i.isupper()]
    low = [i for i in password if i.islower()]
    dig = [i for i in password if i.isdigit()]
    return all([len(password) >= 8, upp, low, dig])
txt = input()
print(is_password_good(txt))
```

```python
def update_dictionary(d, key, value):
    if key in d:
        d[key].append(value)
    elif key is not d:
        if 2*key is d:
            d[2*key].append(value)
        elif (2*key is not d) and d.get(2*key)==None:
            d[2*key]=[]
            d[2*key].append(value)
        elif (2*key is not d) and d.get(2*key)!=None:
            d[2*key].append(value)
    return
```
